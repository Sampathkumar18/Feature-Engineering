```python
#Let's modify the columns:

train_set = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None, index_col=None)
train_set
```

```python
# Let's pick categorical variables alone:

column = [1,3,5,6,7,8,9,13]

train_set = train_set[column]
```

```python
train_set.columns = ["Employment","Degree","Status","Designation","family_Posting","Race","Sex","Country"]
```

```python
train_set
```

```python
train_set.isnull().sum()
```

```python
train_set["Country"].value_counts()
```

```python
#We have" ?" in our dataset, so we need to replace those with " Other" values:

train_set = train_set.replace(" ?", " Others")
```

```python
train_set["Country"].value_counts().index[2]
```

```python
### Using (ORDINAL Number Encoding) to HANDLE CATEGORICAL FEATURE:

Country_map = train_set["Country"].value_counts().to_dict()
Country_map

#So in ORDINAL Number Encoding, we are replacing Categories with that particular Counts.
```

```python
train_set["Country_Counter_M"] = train_set["Country"].map(Country_map)
train_set.head(20)
```

### ADVANTAGES:

1) Easy to implement
2) Not Increasing Feature space

### Disadvantages:

1) it will provide same weight if the frequencies are same
(Eg. United States=12600, Canada=12600)

# (2) TARGET GUIDED ENCODING:

(1) Ordering the label according to the Target


(2) Replace the label by the joint probability of being 1 or 0.

```python
import pandas as pd

df = pd.read_csv("C:\\Users\\mythi\\Downloads\\FEATURE ENGINEERING\\Titanic dataset\\titanic.csv", usecols=["Cabin", "Survived"])
df.head(10)
```

```python
df.isnull().sum()
```

```python
df["Cabin"] = df["Cabin"].fillna("Missing")
```

```python
df
```

```python
#In this thing take a first letter of the category and map them with Numeric based on most frequent occuring(%):
#Lets Get the first letter in cabin feature:

df["Cabin"] = df["Cabin"].astype(str).str[0]
df["Cabin"]
```

```python
df.groupby("Cabin")["Survived"].mean()
```

```python
df.groupby("Cabin")["Survived"].mean().sort_values(ascending=True)
```

```python
ordinal_labels = df.groupby("Cabin")["Survived"].mean().sort_values(ascending=True).index
ordinal_labels
```

```python
ordinal_labels2 = {k:i for i,k in enumerate(ordinal_labels, 0)}
ordinal_labels2
```

```python
df["Cabin_Ordinal_Labels"] = df["Cabin"].map(ordinal_labels2)
```

```python
df.head()
```

```python
# Mean Ordinal Encoding:

mean_ordinal = df.groupby("Cabin")["Survived"].mean().to_dict()
mean_ordinal
```

```python
df["Mean_Ordinal_Encoding"] = df["Cabin"].map(mean_ordinal)
```
