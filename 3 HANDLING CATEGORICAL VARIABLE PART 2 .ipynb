So in this place, many many features are there in each column..
if we use ONE HOT WNCODING TECHNIQUE HERE, IT'LL CREATE A BIG PROBLEM

# So how to fix this kinda of prob?

This kind of problem is solved in one of the kaggle competition and ppl have won the price as well 
using "KDD CUP ORANGE" Technique...

This Technique tells that you can take top 10 categories in every feature of your dataset and proceed ONE HOT ENCODING.



```python
# LET'S IMPLEMENT THIS TECHNIQUE FOR COL "X1" CATEGORIES:
```

```python
df["X1"].value_counts().sort_values(ascending=False) #From this choosing only top 10 categories for ONE HOT ENCODING
```

```python
df["X1"].value_counts().sort_values(ascending=False).head(10)
```

```python
df["X1"].value_counts().sort_values(ascending=False).sum()
```

```python
df["X1"].value_counts().sort_values(ascending=False).head(10).sum()
```

```python
TOP_10Category_Features = (3821/4209)*100
TOP_10Category_Features           #So its pretty reasonnable approach
```

```python
# Okay let's proceed further:

Top10 = df["X1"].value_counts().sort_values(ascending=False).head(10).index
Top10 = list(Top10)
Top10
```

```python
pd.get_dummies(Top10)
```

```python

for categories in Top10:
    df[categories] = np.where(df["X1"]==categories,1,0)
```

```python
df.head(10)
```

```python
Top10.append("X1")
```

```python
#Let's try implementing for all the Features:
```

```python
List = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6']
for i in List:
    CATE = df[i].value_counts().sort_values(ascending=False).head(10).index
    List_Category = list(CATE)
    print(List_Category)
    for categories in List_Category:
        df[categories] = np.where(df[i]==categories,1,0)
```
